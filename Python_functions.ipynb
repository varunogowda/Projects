```python
def fill_values(df):
    for col in df.columns:
        if(col == 'Radius (cm)'):
            a=df.loc[df['Radius (cm)'] == 0]
            cl=a['Fruit (class)']
            cl =cl.values.tolist()
            cl=set(cl)
            cl=list(cl)
            c=df.loc[df['Fruit (class)']==cl[0]]
            me=c['Radius (cm)'].mean()
            me=(round(me,2))
            df['Radius (cm)']=df['Radius (cm)'].replace(0, me)
        elif(col == 'Weight (grams)'):
            a=df.loc[df['Weight (grams)'] == 0]
            cl=a['Fruit (class)']
            cl =cl.values.tolist()
            cl=set(cl)
            cl=list(cl)
            c=df.loc[df['Fruit (class)']==cl[0]]
            me=c['Weight (grams)'].mean()
            me=(round(me,2))
            df['Weight (grams)']=df['Weight (grams)'].replace(0, me)
    return df

fill_values(df)
```

```python
def transform_nominal(data):
    mf= pd.get_dummies(data['Colour'])
    data = pd.concat([data,mf], axis =1).reindex(data.index)
    data.drop('Colour', axis=1, inplace=True)
    return data

transform_nominal(df)
```
